# Método Utilizado

O método KMeans é frequentemente utilizado em tarefas de clusterização para agrupar dados em k grupos distintos com base em características semelhantes. Em relação à identificação de objetos defeituosos, o KMeans pode ser empregado para agrupar itens com características semelhantes, facilitando a identificação de padrões anômalos que podem indicar defeitos.

O algoritmo KMeans funciona da seguinte maneira:

Inicialização dos Centróides: Escolhe-se aleatoriamente k pontos no espaço de dados como os centróides iniciais dos clusters.

Atribuição de Pontos aos Clusters: Cada ponto de dados é atribuído ao cluster cujo centróide é mais próximo.

Atualização dos Centróides: Os centróides de cada cluster são recalculados como a média dos pontos atribuídos a esse cluster.

Repetição: Os passos 2 e 3 são repetidos até que os centróides não mudem significativamente ou até que um número máximo de iterações seja atingido.

O resultado final é uma divisão dos dados em k clusters, com os centróides representando o "centro" de cada cluster.

Em relação a referências bibliográficas, um texto amplamente utilizado para entender o método KMeans é o livro "Introduction to Data Mining"$^{[1]}$ de Tan, Steinbach e Kumar. Nesse livro, os autores explicam detalhadamente o funcionamento do algoritmo KMeans e fornecem exemplos práticos de aplicação.


# Resolução do Problem

O hiperparâmetro K setado foi de 2, para dividir os dados em dois clusters distintos. Inicialmente, os dados são padronizados utilizando o MinMaxScaler para garantir consistência nas escalas das características. Objetos semelhantes devem pertencer ao mesmo cluster, enquanto objetos divergentes podem indicar potenciais defeitos, com isso após utilizar o KMeans, os objetos 1, 3 e 4 pertenceram ao mesmo cluster (grupo 0) já o objeto 2 foi o único que pertence a um cluster diferente (grupo 1), com isso podemos afirmar que aparentemente o objeto potencialmente defeituoso é o Objeto 2.


```python
# Standardize the data
scaler = MinMaxScaler()
scaler.fit_transform(df.loc[:,:])
df=pd.DataFrame(scaler.fit_transform(df.loc[:,:]), columns = df.columns)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=42)  # You may need to adjust the number of clusters based on your data
df['Cluster'] = kmeans.fit_predict(df)

# Display the DataFrame with cluster assignments
print(df)
```
